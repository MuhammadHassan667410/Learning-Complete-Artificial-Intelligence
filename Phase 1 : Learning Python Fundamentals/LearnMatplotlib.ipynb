```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
x = np.arange(0, 6, 1)
y = np.arange(0, 11, 2)
plt.plot(x, y, 'r^:',label='2x1',)
# get a x2 np array that is arranged from 0 to 10
x2  = np.arange(0, 5, 0.5)

#plot half line
plt.plot(x2[:6], x2[:6]**2, 'go--',label='x2^2')
# plot remaining line
plt.plot(x2[5:], x2[5:]**2, 'b*:',label='x2^2')
plt.title('My First Graph', fontdict={'fontsize':20, 'color':'red','fontweight':'bold'})
plt.xlabel('x-axis')
plt.xticks([0,1,2,3,4,5,6])
plt.ylabel('y-axis')
#plt.yticks([0,2,4,6,8,10,12])
plt.legend(loc='upper left', labelcolor='orange')
# save tge figure
plt.savefig('my_first_graph.png')
plt.show()
```

```python
labels = ['A','B','C']
values = [1,4,2]
bars = plt.bar(labels,values)
plt.title('Bar Graph', fontdict={'fontsize':20, 'color':'red', 'fontweight':'bold'})
plt.xlabel('x-axis')
plt.ylabel('y-axis')
bars[0].set_hatch('|')
bars[1].set_hatch('o')
bars[2].set_hatch('*')
plt.figure(figsize=(3,2))
plt.show()
```

```python
fifa = pd.read_csv('fifa_data.csv')
fifa.head()
```

```python
bins = np.arange(40, 110, 10)
plt.hist(fifa.Overall, bins=bins, color='r')
plt.title(' Distribution of football players in 2018 ', fontweight='bold', fontsize=15)
plt.xlabel('Overall Rating', fontsize=13)
plt.xticks(bins)
plt.ylabel('Number of players', fontsize=13)
plt.show()
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]
#left
#right
labels = ['Left', 'Right']
colors = ['b', 'r']
plt.title('Foot preferrece of Fifa Players', fontweight='bold', fontsize=15)
plt.pie([left, right], labels = labels, colors = colors, autopct='%.4f')
plt.show()
```

```python
# Check empty rows in Weight and print the empty rows
print(fifa.Weight.isnull().sum())
fifa.loc[fifa.Weight.isnull()]
```

```python
# Replace 'lbs' in Weight with ''
fifa.Weight = fifa.Weight.replace('lbs', '').astype('float')
#fifa.Weight
# Change dtype to int64 ignoring missing values
fifa.Weight = pd.to_numeric(fifa.Weight, errors='coerce').astype('Int64')
fifa.Weight
```

```python

# players under 125
light = fifa.loc[fifa.Weight < 125].count()[0]
# players over 125 and under 150
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
# players over 150 and under 175
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight <= 175)].count()[0]
# players over 175 and under 200
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight <= 200)].count()[0]
# players over 200
heavy = fifa.loc[fifa.Weight > 200].count()[0]
weight = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']
explode = (.4,.1,0,0,.4)
plt.pie(weight, labels=labels, autopct='%.2f', pctdistance=0.9, explode=explode)
# Change style of the pie chart
plt.style.use('ggplot')
# solve numbers over each other problem
plt.axis('equal')

plt.title('Weight of Fifa Players', fontweight='bold', fontsize=15)
plt.show()
```

```python
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
juventus = fifa.loc[fifa.Club == 'Juventus']['Overall']
labels = ['FC Barcelona', 'Real Madrid', 'Juventus']
plt.figure(figsize=(5,7))
plt.style.use('default')
boxes = plt.boxplot([barcelona, madrid, juventus], labels = labels, patch_artist=True)
# Add colors in the boxes
for box in boxes['boxes']:
    box.set(color='r', linewidth=2)
    box.set(facecolor='b')
plt.title('Comparision of football clubs', fontweight='bold', fontsize=15)
plt.ylabel('Overall Rating')
plt.xlabel('Football Clubs')
plt.show()
```
