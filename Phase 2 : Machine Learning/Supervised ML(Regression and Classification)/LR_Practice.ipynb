```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('mouse_weight_size.csv')
df
```

```python
y = df['Mouse_Size']
x = df[['Mouse_Weight']]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x, y)
```

```python
y_pred = lr.predict(x)
y_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
y_pred_mse = mean_squared_error(y, y_pred)
y_pred_mse
```

```python
y_pred_r2 = r2_score(y, y_pred)
y_pred_r2
```

```python
import matplotlib.pyplot as plt

plt.scatter(x, y, color='blue')
plt.plot(x, y_pred, c='r')
plt.xlabel("Mouse Weight")
plt.ylabel("Mouse Size")
plt.title("Mouse Size vs Weight")
plt.show()
```
