# Load Data

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('delaney_solubility_with_descriptors.csv')
```

```python
df
```

# Split Data into X and Y

```python
y = df['logS']
y
```

```python
x = df.drop('logS', axis=1)
x
```

# Split Data into Training and testing sets

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100)
```

```python
x_test
```

# Building Model

## Linear Regression

### Training model

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

### Appplying the model to make prediction

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

### Evaluate Model Perfomence

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
lr_train_mse
```

```python
lr_train_r2
```

```python
lr_test_mse
```

```python
lr_test_r2
```

```python
lr_results = pd.DataFrame(['Linear Regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ['Method', 'Train MSE', 'Train R2', 'Test MSE', 'Test R2']
```

```python
lr_results
```

## Random Forest

### Training model

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(max_depth=2, random_state=100)
```

```python
rf.fit(x_train, y_train)
```

### Applying the model to make prediction

```python
y_rf_train_pred = rf.predict(x_train)
y_rf_test_pred = rf.predict(x_test)
```

```python
y_rf_train_pred
```

```python
y_train
```

```python
y_test
```

```python
y_rf_test_pred
```

### Evaluate Model Perfomeence

```python
from sklearn.metrics import mean_squared_error, r2_score
y_train_mse = mean_squared_error(y_train, y_rf_train_pred)
y_train_r2 = r2_score(y_train, y_rf_train_pred)

y_test_mse = mean_squared_error(y_test, y_rf_test_pred)
y_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
rf_results = pd.DataFrame(['Random Forest Regressor', y_train_mse, y_train_r2, y_test_mse, y_test_r2]).transpose()
rf_results.columns = ['Method', 'Train MSE', 'Train R2', 'Test MSE', 'Test R2']
rf_results
```

## Model Comparison

```python
df_models = pd.concat([lr_results, rf_results], axis=0)
```

```python
df_models
```

```python
df_models.reset_index(drop=True)
```
